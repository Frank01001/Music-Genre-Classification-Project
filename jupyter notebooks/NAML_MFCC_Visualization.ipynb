```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = './mfcc_dataset.csv'

data = pd.read_csv(path)
data.describe()
```

```python
corr = data.corr()
```

```python
fig, ax = plt.subplots(figsize=(40,36))
sns.heatmap(corr, annot = True, cmap = 'vlag_r', vmin = -1, vmax = 1, ax=ax)
```
